## NumPy Array (Core Concept)
- NumPy is the most popular Python library for numerical computing.
- Fast (written in C)

- Works with arrays & matrices
- Supports math, statistics, and linear algebra

```python
import numpy as np

a = np.array([1, 2, 3])
print(a)
```

```python
b = np.array([[1,2,3],[4,5,6]])
print(b)
```

## Useful Array Functions

```python
arr = np.array([1,2,3,4,5])

print(arr.shape)     # size
print(arr.dtype)     # data type
print(arr.sum())     # total
print(arr.mean())    # average
print(arr.max())     # max
print(arr.min())     # min
```

## Creating Special Arrays 

```python
np.zeros((2,3))      # all zeros
np.ones((3,3))       # all ones
np.arange(1,10)      # 1 to 9
np.linspace(0,1,5)   # 5 values between 0 & 1
```

## Array Math (Element-wise)

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
print(a * b)
print(a ** 2)
```

## Matrix Operations (Important for ML)

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print(np.dot(A,B))       # matrix multiplication
print(np.linalg.inv(A))  # inverse
print(np.linalg.det(A))  # determinant
```

## Basic Example Line Chart

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [2, 5, 7, 8]

plt.plot(x, y)
plt.title("Simple Line Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

## Bar Chart Example

```python
import matplotlib.pyplot as plt

marks = [70, 80, 100, 90]
subjects = ["science", "math", "social", "english"]

plt.bar(marks, subjects)
plt.title("Student Marks")
plt.xlabel("marks")
plt.ylabel("subjects")
plt.show()
```

## Scatter Plot Example

```python
import matplotlib.pyplot as plt

height = [150, 160, 170, 180]
weight = [50, 60, 70, 80]

plt.scatter(height, weight)
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (KG)")
plt.show()
```

## Outlier Removal Using Z Score

```python
data = pd.read_csv("C:\Arun\Data/train_loan.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
sns.boxplot(x="CoapplicantIncome", data=data)
plt.show()
```

```python
sns.histplot(x="CoapplicantIncome", data=data)
plt.show()
```

```python
main_range = data["CoapplicantIncome"].mean() - (3*data["CoapplicantIncome"].std())
max_range = min_range = data["CoapplicantIncome"].mean() + (3*data["CoapplicantIncome"].std())
```

```python
main_range, max_range
```

```python
new_data=data[data["CoapplicantIncome"]<=max_range]
```

```python
sns.boxplot(x="CoapplicantIncome", data=new_data)
```

```python
z_score = (data["CoapplicantIncome"] - data["CoapplicantIncome"].mean())/(data["CoapplicantIncome"].std())
```

```python
z_score
```

```python
data["z_score"] = z_score
```
