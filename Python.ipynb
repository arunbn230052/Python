```python
def student_marks_system():
    name = input("Enter student name: ")

    marks = {}
    subjects = ["Math", "Science", "English"]

    for subject in subjects:
        marks[subject] = int(input(f"Enter marks for {subject}: "))

    total = sum(marks.values())
    percentage = total / len(subjects)

    if percentage >= 90:
        grade = "A+"
    elif percentage >= 80:
        grade = "A"
    elif percentage >= 70:
        grade = "B"
    elif percentage >= 60:
        grade = "C"
    else:
        grade = "Fail"

    print("\n--- Student Report ---")
    print("Name:", name)
    print("Marks:", marks)
    print("Total Marks:", total)
    print("Percentage:", percentage)
    print("Grade:", grade)


student_marks_system()
```

## Class and Object
- A class is a blueprint for creating objects. It defines attributes (variables) and methods (functions).
- An object is an instance of a class. Each object can have its own values for attributes.

```python
# Class definition
class Student:
    def __init__(self, name, age):
        self.name = name  # attribute
        self.age = age    # attribute

    def display(self):    # method
        print(f"Name: {self.name}, Age: {self.age}")

# Object creation
student1 = Student("Arun", 21)
student2 = Student("Sita", 22)

# Accessing method
student1.display()
student2.display()
```

## Inheritance
- Inheritance allows a class (child) inherit attribute and method from another class (parient).

```python
# Parent class
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, my name is {self.name}")

# Child class
class Student(Person):
    def __init__(self, name, student_id):
        super().__init__(name)  # inherit name from Person
        self.student_id = student_id

    def show_id(self):
        print(f"My Student ID is {self.student_id}")

# Object creation
s = Student("Arun", 101)
s.greet()       # inherited method
s.show_id()     # child method
```

## Eigenvalues & Eigenvectors

<center/>Av=λv

## Bayes Theorem
- Bayes’ Theorem describes how to update the probability of an event based on new evidence.

```python
P_D = 0.01
P_pos_given_D = 0.99
P_pos_given_notD = 0.01

P_pos = (P_pos_given_D * P_D) + (P_pos_given_notD * (1 - P_D))

P_D_given_pos = (P_pos_given_D * P_D) / P_pos
print(P_D_given_pos)
```

## Calculus
- Calculus is a branch of mathematics that deals with rates of change and accumulation.
- A derivative measures how fast a value changes.

```python
import sympy as sp
x = sp.symbols('x')
f = x**2                                #    dy/dx
df = sp.diff(f, x)                      # f(x,y)=x2+y2
print(df)
```

```python
# An integral represents area under the curve.
integral = sp.integrate(f, x)
print(integral)                       
```

```python
# Limits
import sympy as sp

x = sp.symbols('x')
f = (x**2 - 4) / (x - 2)
limit = sp.limit(f, x, 2)
print(limit)
```

```python
import sympy as sp

x = sp.symbols('x')
f = x**3 + 2*x
df = sp.diff(f, x)
print(df)
```

```python
# A partial derivative measures how a function changes with respect to one variable, while keeping other variables constant.
import sympy as sp

x, y = sp.symbols('x y')
f = x**2 * y + y**3

dx = sp.diff(f, x)
dy = sp.diff(f, y)

print("∂f/∂x =", dx)
print("∂f/∂y =", dy)
```

```python
# Gradient Descent is an optimization algorithm used to minimize a function (usually a loss/cost function) by moving in the opposite direction of the gradient.
x = 10
lr = 0.1

for i in range(10):
    grad = 2 * x
    x = x - lr * grad
    print(x)
```
